```python
# preparing data for upload to psql
import pandas as pd
pd.options.display.max_columns = 100
pd.options.display.max_rows = 9999
```

```python
# Prepare impact area study data
ag_vul_to_cc = pd.read_excel('./ncomms6989-s2.xlsx', skiprows=3)
ag_vul_to_cc.drop(['Unnamed: 0', 'Unnamed: 4', 'Unnamed: 7', 'Unnamed: 10'], axis=1, inplace=True)
ag_vul_to_cc.drop([145,146], axis=0, inplace=True)
ag_vul_to_cc.columns = ['country', 'maize_per_area_sig', 'maize_avg_impact_on_sig_area', 
                       'rice_per_area_sig', 'rice_avg_impact_on_sig_area',
                       'wheat_per_area_sig', 'wheat_avg_impact_on_sig_area',
                       'soy_per_area_sig', 'soy_avg_impact_on_sig_area']
ag_vul_to_cc.to_csv("ag_vul_to_cc.csv")
ag_vul_to_cc
```

```python
# Prepare allocating aid study data
alloc_aid = pd.read_excel('./Quantifying_Vulnerability_DB.xls', skiprows=2)
alloc_aid.columns = ['country', 
                    'oa_clim_dri', 'oa_clim_vuln_adj_inc_reg', 'oa_proj_concern_adj_cost_prob_succ',
                    'oa_pop_2008', 'oa_clim_dri_ind_aid_share', 'oa_vuln_ind_aid_share', 'oa_proj_concern_ind_aid_share',
                    'ew_clim_dri', 'ew_clim_vuln_adj_inc_reg', 'ew_proj_concern_adj_cost_prob_succ',
                    'ew_pop_2008', 'ew_clim_dri_ind_aid_share', 'ew_vuln_ind_aid_share', 'ew_proj_concern_ind_aid_share',
                    'slr_clim_dri', 'slr_clim_vuln_adj_inc_reg', 'slr_proj_concern_adj_cost_prob_succ',
                    'slr_storm_surge_pop_2008', 'slr_clim_dri_ind_aid_share', 'slr_vuln_ind_aid_share', 'slr_proj_concern_ind_aid_share',
                    'apl_clim_dri', 'apl_clim_vuln_adj_inc_reg', 'apl_proj_concern_adj_cost_prob_succ',
                    'apl_rural_pop_2008', 'apl_clim_dri_ind_aid_share', 'apl_vuln_ind_aid_share', 'apl_proj_concern_ind_aid_share',
                    'income_per_cap_2008', 'kkm_reg_qual_2008', 'area', 'island', 'small_island',
                    'world_region', 'world_bank_region', 'world_sub_region', 'iso3', 'income_status', 'world_bank_lending_class']
alloc_aid
# May have to iterate through and replace NaN / na with -9999 or some other indicator
# for postgres to be happy
```

```python
# Add data to postgres
# Tutorial for psycopg2: http://www.postgresqltutorial.com/postgresql-python/insert/
import psycopg2
import os

print(os.environ["DATABASE_URL"])

connection = psycopg2.connect(database="nathansuberi", user="nathansuberi")
cursor = connection.cursor()



cursor.execute("DROP TABLE IF EXISTS ag_vul_to_cc")
cursor.execute("CREATE TABLE ag_vul_to_cc (id SERIAL PRIMARY KEY, country VARCHAR NOT NULL," + 
                       "maize_per_area_sig FLOAT, maize_avg_impact_on_sig_area FLOAT, "+
                       "rice_per_area_sig FLOAT, rice_avg_impact_on_sig_area FLOAT, " + 
                       "wheat_per_area_sig FLOAT, wheat_avg_impact_on_sig_area FLOAT, " +
                       "soy_per_area_sig FLOAT, soy_avg_impact_on_sig_area FLOAT);")
connection.commit()

cursor.execute("SELECT table_name FROM information_schema.tables WHERE table_schema = 'public';")
print(cursor.fetchall())
connection.commit()


```
